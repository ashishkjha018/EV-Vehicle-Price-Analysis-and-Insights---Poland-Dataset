```python
# Task 2 - Outliers in "mean - Energy consumption [kWh/100 km]"

# Loading the Excel (.xlsx) file here
df= pd.read_excel("FEV-data-Excel.xlsx")

energy_column = 'mean - Energy consumption [kWh/100 km]'

# Check for missing values and remove them if necessary
df = df.dropna(subset=[energy_column])

# Calculate the mean and standard deviation
mean = df[energy_column].mean()
std_dev = df[energy_column].std()

threshold = 1.5  # We can take upto 2, 2 and 3 both are good thresholds for finding anomalies but here the values are very close to mean so we take 1.5

outliers = df[(df[energy_column] < mean - threshold * std_dev) | 
              (df[energy_column] > mean + threshold * std_dev)]

print(f"Outliers in Energy Consumption [kWh/100 km] with Threshold: {threshold}:")
print(outliers)
```

```python
# Task 3(a) - Scatter plot to visualise relationship between battery capacity and range.

import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Excel (.xlsx) file here
df= pd.read_excel("FEV-data-Excel.xlsx")

battery_capacity_column = 'Battery capacity [kWh]'
range_column = 'Range (WLTP) [km]'

# Drop rows with missing values in either column
df_clean = df.dropna(subset=[battery_capacity_column, range_column])

# Using Scatter plot to find similarities
plt.figure(figsize=(10, 6))
sns.scatterplot(x=battery_capacity_column, y=range_column, data=df_clean)

plt.title('Battery Capacity vs. Range of EVs')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Range (km)')

plt.show()

# Calculating correlation coefficient to check strength between these columns

correlation = df_clean[battery_capacity_column].corr(df_clean[range_column])
print(f"Correlation coefficient between Battery Capacity and Range: {correlation}")
```

```python
# Task 4 - Build an EV recommendation class

# Loading the Excel (.xlsx) file here
df= pd.read_excel("FEV-data-Excel.xlsx")

class EVRecommendation:
    def __init__(self, df):
        self.df = df  # Storing DataFrame for access in methods
    
    def recommend(self, budget, min_range, min_battery_capacity):
        """
        Recommends the top 3 EVs based on the user's budget, desired range, and battery capacity.
        
        Parameters:
        - budget: The user's maximum budget.
        - min_range: The minimum desired range.
        - min_battery_capacity: The minimum desired battery capacity.
        
        Returns:
        - A list of top 3 recommended EVs matching the criteria.
        """
        filtered_df = self.df[
            (self.df['Minimal price (gross) [PLN]'] <= budget) & 
            (self.df['Range (WLTP) [km]'] >= min_range) & 
            (self.df['Battery capacity [kWh]'] >= min_battery_capacity)
        ]
        
        if filtered_df.empty:
            return "No EVs match the given criteria."
        
        sorted_df = filtered_df.sort_values(by=['Range (WLTP) [km]', 'Battery capacity [kWh]'], ascending=False)

        top_3_evs = sorted_df.head(3)
        return top_3_evs[['Make', 'Model', 'Minimal price (gross) [PLN]', 'Range (WLTP) [km]', 'Battery capacity [kWh]']]

#Trying our Class for output
ev_recommender = EVRecommendation(df)

budget = 380000  
min_range = 400 
min_battery_capacity = 70  

top_3 = ev_recommender.recommend(budget, min_range, min_battery_capacity)
print(top_3)
```

```python
# Task 5 - Inferential Statistics – Hypothesis Testing

from scipy.stats import ttest_ind

# Loading the Excel (.xlsx) file here
df= pd.read_excel("FEV-data-Excel.xlsx")

print(df.columns)

tesla_data = df[df['Make'] == 'tesla']
audi_data = df[df['Make'] == 'Audi']

tesla_engine_power = tesla_data['Engine power [KM]'].dropna()
audi_engine_power = audi_data['Engine power [KM]'].dropna()

print(f"Tesla Standard Deviation: {tesla_engine_power.std()}")
print(f"Audi Standard Deviation: {audi_engine_power.std()}")

t_stat, p_value = ttest_ind(tesla_engine_power, audi_engine_power)

alpha = 0.05  # significance level

if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Failed to reject the null hypothesis"

# Printing Insights and Conclusion
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")
print(f"Based on the p-value, we {result} and conclude that there is a significant difference in engine power between Tesla and Audi.")

# Printing actionable insights and recommendations
if result == "Reject the null hypothesis":
    print("Insight: The analysis suggests that there is a significant difference in engine power between Tesla and Audi. This may imply different performance characteristics or design philosophies between the two manufacturers.")
else:
    print("Insight: The analysis suggests that there is no significant difference in engine power between Tesla and Audi. Both manufacturers seem to offer comparable engine power in their EVs.")
```
